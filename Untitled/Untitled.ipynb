```python
from time import sleep
from io import BytesIO
from random import choice
screen = W.Image()
display(screen)
display(txt)
key = 0
def on_click(z):
    global key
    txt.value=str(z)
    key = int(z.description)
buttons = [W.Button(description =str(i)) for i in range(10)]
for i in range(10):
    buttons[i].on_click(on_click)
display(W.HBox(buttons))
def cb(z):
    global key
    env.step(key)
    key = 0
    with BytesIO() as bio:
        Image.fromarray(env.render(mode='rgb_array')).save(bio, 'png')        
        screen.value = bio.getvalue()
p = W.Play(min=0, max =10000)
p.observe(cb)
display(p)
```
