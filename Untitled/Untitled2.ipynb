```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')
x = np.linspace(-1,1, 20)
plt.plot(x, x**2,linewidth=12,color=(0.7,0.9,0.2), markerfacecolor='blue',
        linestyle='-', marker='p', markersize=16, label='y=x^2')
plt.plot(x, np.sin(x),linewidth=4,color=(0.1,0.5,1), markerfacecolor='green',
        linestyle='-', marker='p', markersize=8, label='y=sin(x)')
plt.legend()
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')
color = (0.5, 1., 0.9)
t = np.linspace(0, 2*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), '--', color=color, linewidth=1.5)
plt.plot(.7*np.cos(t), .7*np.sin(t), color=color, linewidth=1.5)
plt.plot(1.3*np.cos(t), 1.3*np.sin(t), color=color, linewidth=1.5)

for i in range(8):
    t0 = 2*np.pi/8*i
    x0,y0 = 1.15*np.cos(t0), 1.15*np.sin(t0)
    plt.plot(.15*np.cos(t)+x0, .15*np.sin(t)+y0,  color=color, linewidth=1.5)

    

t = np.linspace(0, 4*np.pi, 6)
plt.plot(.7*np.cos(t+np.pi/10), .7*np.sin(t+np.pi/10),  color=color, linewidth=1.5)


t = np.linspace(0, 2*np.pi, 5)
plt.plot(np.cos(t), np.sin(t), ':' , color=color, linewidth=1.5)
plt.plot(np.cos(t+np.pi/4), np.sin(t+np.pi/4), ':', color=color, linewidth=1.5)

TEXT="12345678"
for i in range(8):
    s = 2*np.pi * (4-i) /8
    x0, y0 = 1.12*np.cos(s), 1.12*np.sin(s)
    plt.text(x0, y0, TEXT[i], color=color, 
             ha='center', va='center', fontsize=20, 
             rotation=s*180/np.pi-90)
    
    
    
TEXT="ksl"
for i in range(len(TEXT)):
    s = 2*np.pi * (-i) /len(TEXT)
    x0, y0 = .75*np.cos(s), .75*np.sin(s)
    plt.text(x0, y0, TEXT[i], color='w', 
             ha='center', va='center', fontsize=16, 
             rotation=s*180/np.pi-90)
    
plt.savefig('circle.png')
```
