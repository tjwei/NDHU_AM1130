```
class 動物:
    "這是我們課堂上介紹 class 的範例 class"
    def __init__(self, name):
        self.name = name
        self.暱稱 = "~~~"+name+"~~~~"
        self.體力 = 10
        self.物種 = '動物'
    def 叫(self):
        if self.體力 <0:
          print(self.name, "慘叫了一聲")
        else:
          print(self.name, "叫了一聲")
    def 跑(self):
        if self.體力 <0:
          print(self.name, "跑不動了")
        else:          
          print(self.name, "跑了一下")
          self.體力 -= 3
    def 飛(self):
        print(self.name, "不會飛")
    def __str__(self):
      return self.暱稱
    def __repr__(self):
      return "<{}: {}>".format(self.物種, self.name)
```

```
class 狗(動物):
  def __init__(self, name):
    super().__init__(name)
    self.物種 = "狗"
    self.體力 = 35
    
  def 叫(self):
    if self.體力 <0:
        print(self.name, "慘叫了一聲嗚嗚")
    else:
        print(self.name, "叫了一聲汪汪")
  
```

```
class 貓(動物):
  def __init__(self, name):
    super().__init__(name)
    self.物種 = "貓"
    self.體力 = 30
  def 叫(self):
    if self.體力 <0:
        print(self.name, "慘叫了一聲喵嗚嗚")
    else:
        print(self.name, "叫了一聲喵喵")
```

```
class 鳥(動物):
  def __init__(self, name):
    super().__init__(name)
    self.物種 = "鳥"
  def 叫(self):
    if self.體力 <0:
        print(self.name, "慘叫了一聲鳩烏烏")
    else:
        print(self.name, "叫了一聲啾啾")
  def 飛(self):
    if self.體力 < 0:
      print(self.name, "飛不動了")
    else:
      print(self.name, "飛了起來")
      self.體力 -= 5
```

```
旺福 = 狗('旺福')
喵喵 = 貓('喵喵')
bird = 鳥('bird')
```

```
from random import choice
zoo = [狗('旺福'), 貓('喵喵'), 鳥('bird')]
for i in range(1):
  animal = choice(zoo)
  print("------{}-----".format(animal))
  animal.飛()
  animal.跑()
  animal.叫()
```
